```python
from GeneralGlobalModel_Class import GlobalModel
import Reactions_H as R
import matplotlib.pyplot as plt
import numpy as np
me=9.11e-31
mp=1.67e-27
q=1.6e-19
reactions=R.reactions1
gm1=GlobalModel(reactions,[2*mp,0],Tg=200)
gm2=GlobalModel(reactions,[2*mp,0],Tg=300)
gm3=GlobalModel(reactions,[2*mp,0],Tg=600)
```

```python
'''No he definido esta funcin dentro de la clase GlobalModel, por que necesita
mucho tiempo en calcular la solucin 2D. Por eso, calculo el resultado
con gm.Find_solution_2D_sweep y con la funcin Zonas() represento los mayoritarios utilizando la
solucin anterior'''
def zonas(n1array,n2array,n3array):
    N_mayoritario=np.empty((n1array.shape[0],n1array.shape[1]))
    N_mayoritario[:]=np.nan
    limites=np.empty((n1array.shape[0],n1array.shape[1]))
    limites[:]=np.nan
    for i in range (n1array.shape[0]):
        for j in range (n1array.shape[1]):
            if (n1array[i,j]>n2array[i,j])&(n1array[i,j]>n3array[i,j]):
                N_mayoritario[i,j]=1
            if (n2array[i,j]>n1array[i,j])&(n2array[i,j]>n3array[i,j]):
                N_mayoritario[i,j]=2
            if (n3array[i,j]>n2array[i,j])&(n3array[i,j]>n1array[i,j]):
                N_mayoritario[i,j]=3
    for i in range(N_mayoritario.shape[0]-1):
        for j in range(N_mayoritario.shape[1]-1):
            if (N_mayoritario[i,j]!=N_mayoritario[i+1,j])|(N_mayoritario[i,j]!=N_mayoritario[i,j+1]):
                limites[i,j]=1
    return [N_mayoritario,limites]

pmin,pmax=1e5,6e6
nmin,nmax=4e13,7e14
Nn,Np=80,80
nn,pp,[ne,nH,n1,n2,n3,e]=gm1.Find_solution_2D_sweep(pmin,pmax,nmin,nmax,Np,Nn,scale='log')
N_mayoritario1,limites1=zonas(n1,n2,n3)
nn,pp,[ne,nH,n1,n2,n3,e]=gm2.Find_solution_2D_sweep(pmin,pmax,nmin,nmax,Np,Nn,scale='log')
N_mayoritario2,limites2=zonas(n1,n2,n3)
nn,pp,[ne,nH,n1,n2,n3,e]=gm3.Find_solution_2D_sweep(pmin,pmax,nmin,nmax,Np,Nn,scale='log')
N_mayoritario3,limites3=zonas(n1,n2,n3)
```

```python
plt.figure(figsize=(8,4))
plt.scatter((nn+nH).flatten(),pp.flatten(),c=limites1.flatten().T,marker='x',label='$T_i$={}'.format(gm1.Ti))
plt.scatter((nn+nH).flatten(),pp.flatten(),c=limites2.flatten().T,marker='+',label='$T_i$={}'.format(gm2.Ti))
plt.scatter((nn+nH).flatten(),pp.flatten(),c=limites3.flatten().T,marker='.',label='$T_i$={}'.format(gm3.Ti))
plt.xlim([nmin,nmax])
plt.ylim([pmin,pmax])
plt.xscale('log')
plt.yscale('log')
#plt.legend()
plt.xlabel('$n_0$ (cm$^{-3}$)')
plt.ylabel('$P_\mu$ (W/m$^{-3}$)')
plt.savefig('Figures\sensibilidad_Tg_predominance.pdf')
```
