```python
me=9.11e-31
mp=1.67e-27
q=1.6e-19

def Reaction_rate(d,e):
    s=0
    for i in range (len(d)):
        s=s+d[i]*np.log(e)**i
    return np.exp(s)

a3rr=[-1.78e1,2.278,-1.2668,4.296e-1,-9.6099e-2,1.3879e-2,-1.231e-3,6.04e-5,-1.247e-6]
a3=['e+H2+=>e+H++H','dissociation',[[me,-1,1],[2*mp,1,1]],[[me,-1,1],[mp,1,1],[mp,0,1]],2.7,lambda x:Reaction_rate(a3rr,x)]

def Reaction1(e):
    return 1.0702e-7*e**0.004876*np.exp(-9.69028/e)
```
