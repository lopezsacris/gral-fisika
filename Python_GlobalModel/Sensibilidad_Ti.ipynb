```python
'''No he definido esta funcin dentro de la clase GlobalModel, por que necesita
mucho tiempo en calcular la solucin 2D. Por eso, calculo el resultado
con gm.Find_solution_2D_sweep y con la funcin Zonas() represento los mayoritarios utilizando la
solucin anterior'''
def zonas(n1array,n2array,n3array):
    N_mayoritario=np.empty((n1array.shape[0],n1array.shape[1]))
    N_mayoritario[:]=np.nan
    limites=np.empty((n1array.shape[0],n1array.shape[1]))
    limites[:]=np.nan
    for i in range (n1array.shape[0]):
        for j in range (n1array.shape[1]):
            if (n1array[i,j]>n2array[i,j])&(n1array[i,j]>n3array[i,j]):
                N_mayoritario[i,j]=1
            if (n2array[i,j]>n1array[i,j])&(n2array[i,j]>n3array[i,j]):
                N_mayoritario[i,j]=2
            if (n3array[i,j]>n2array[i,j])&(n3array[i,j]>n1array[i,j]):
                N_mayoritario[i,j]=3
    for i in range(N_mayoritario.shape[0]-1):
        for j in range(N_mayoritario.shape[1]-1):
            if (N_mayoritario[i,j]!=N_mayoritario[i+1,j])|(N_mayoritario[i,j]!=N_mayoritario[i,j+1]):
                limites[i,j]=1
    return [N_mayoritario,limites]
def to_vector(nn,pp,limites):
    I,J=nn.shape
    n_lims,p_lims=[],[]
    for i in range (I):
        for j in range (J):
            n=nn[i,j]
            p=pp[i,j]
            if limites[i,j]==1:
                n_lims.append(n)
                p_lims.append(p)
    return (n_lims,p_lims)

pmin,pmax=1e5,6e6
nmin,nmax=4e13,4.5e14
Nn,Np=100,100
nn,pp,[ne1,nH1,n11,n21,n31,e1]=gm1.Find_solution_2D_sweep(pmin,pmax,nmin,nmax,Np,Nn,scale='log')
N_mayoritario1,limites1=zonas(n11,n21,n31)
nn,pp,[ne2,nH2,n12,n22,n32,e2]=gm2.Find_solution_2D_sweep(pmin,pmax,nmin,nmax,Np,Nn,scale='log')
N_mayoritario2,limites2=zonas(n12,n22,n32)
nn,pp,[ne3,nH3,n13,n23,n33,e3]=gm3.Find_solution_2D_sweep(pmin,pmax,nmin,nmax,Np,Nn,scale='log')
N_mayoritario3,limites3=zonas(n13,n23,n33)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(nn.flatten(),pp.flatten(),c=limites1.flatten().T,marker='x')
plt.scatter(nn.flatten(),pp.flatten(),c=limites2.flatten().T,marker='+')
plt.scatter(nn.flatten(),pp.flatten(),c=limites3.flatten().T,marker='.')
plt.xlim([nmin,nmax])
plt.ylim([pmin,pmax])
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$n_0$ (cm$^{-3}$)')
plt.ylabel('$P_\mu$ (W/m$^{-3}$)')
plt.savefig('sensibilidad_Ti_predominance.pdf')
```
