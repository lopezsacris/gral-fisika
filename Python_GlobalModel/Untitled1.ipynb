```python
from GeneralGlobalModel_Class import GlobalModel
import Reactions_He as RHe
import Reactions as R
import numpy as np
import matplotlib.pyplot as plt
me=9.11e-31
mp=1.67e-27
q=1.6e-19
```

```python
reactionsHe=RHe.reactions
reactionsH=R.reactions1
gmHe=GlobalModel(reactionsHe,[4*mp,0])
gmH=GlobalModel(reactionsH,[2*mp,0])
```

```python
n0=1e15
pmin,pmax=3e5,3e6
p0=3e6
nmin,nmax=1e14,2e15
N=30
fig,ax=plt.subplots(1,2,figsize=(10,3))
ps,names0,values0=gmH.Contributions_p_sweep([mp,0],n0,pmin,pmax,N)
for i in range (values0[0,:].size):
    ax[1].plot(ps,abs(values0[:,i]),label=names0[i])
ns,names0,values0=gmH.Contributions_n_sweep([mp,0],p0,nmin,nmax,N)
for i in range (values0[0,:].size):
    ax[0].plot(ns,abs(values0[:,i]),label=names0[i])
ax[1].set_xlabel('P$_\mu$(W/m$^3$)')
ax[1].set_ylabel('$n_in_j$|<$\sigma$v>|')
ax[0].set_xlabel('$n_0$ (cm$^{-3}$)')
ax[0].set_ylabel('$n_in_j$|<$\sigma$v>|')
ax[1].set_yscale('log')
ax[0].set_yscale('log')
ax[1].legend(bbox_to_anchor=(1.04,1), loc="upper left")
#fig.suptitle('Reactions involving H', fontsize=16)
plt.tight_layout()
```
