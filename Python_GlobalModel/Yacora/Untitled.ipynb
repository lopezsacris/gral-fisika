```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f=open('YacoraRun_PopCoeff_B1Sigma.dat','r')
data=np.array([0,0,0])
for line in f:
    if (len(line.split())>0 and line.split()[0]!='#'):
        line_array=np.array(line.split(),dtype=float)
        data=np.vstack((data,line_array))
data=np.delete(data,0,0)
```

```python
plt.scatter(data[:,0],data[:,1],c=data[:,2])
plt.xscale('log')
plt.yscale('log')
plt.colorbar()
```
