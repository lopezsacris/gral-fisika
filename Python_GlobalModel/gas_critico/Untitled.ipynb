```python
import pandas as pd
import matplotlib.pyplot as plt
data_01=pd.read_csv("test_0_1_tab.txt",delimiter="\t")
data_01.drop(data_01.tail(1).index,inplace=True) 
data_01[' Pressure (mBar)']=data_01[' Pressure (mBar)'].apply(lambda x: x.replace(',','.')).astype(float)
data_02=pd.read_csv("test_0_2_tab.txt",delimiter="\t")
data_02.drop(data_02.tail(1).index,inplace=True) 
data_02[' Pressure (mBar)']=data_02[' Pressure (mBar)'].apply(lambda x: x.replace(',','.')).astype(float)
data_03=pd.read_csv("test_0_3_tab.txt",delimiter="\t")
data_03.drop(data_03.tail(1).index,inplace=True)
data_03[' Pressure (mBar)']=data_03[' Pressure (mBar)'].apply(lambda x: x.replace(',','.')).astype(float)
data_11=pd.read_csv("test_-3_1_tab.txt",delimiter="\t")
data_11.drop(data_11.tail(1).index,inplace=True) 
data_11[' Pressure (mBar)']=data_11[' Pressure (mBar)'].apply(lambda x: x.replace(',','.')).astype(float)
data_12=pd.read_csv("test_-3_2_tab.txt",delimiter="\t")
data_12.drop(data_12.tail(1).index,inplace=True) 
data_12[' Pressure (mBar)']=data_12[' Pressure (mBar)'].apply(lambda x: x.replace(',','.')).astype(float)
data_13=pd.read_csv("test_-3_3_tab.txt",delimiter="\t")
data_13.drop(data_13.tail(1).index,inplace=True) 
data_13[' Pressure (mBar)']=data_13[' Pressure (mBar)'].apply(lambda x: x.replace(',','.')).astype(float)
data_21=pd.read_csv("test_-6_1_tab.txt",delimiter="\t")
data_21.drop(data_21.tail(1).index,inplace=True) 
data_21[' Pressure (mBar)']=data_21[' Pressure (mBar)'].apply(lambda x: x.replace(',','.')).astype(float)
data_22=pd.read_csv("test_-6_2_tab.txt",delimiter="\t")
data_22.drop(data_22.tail(1).index,inplace=True) 
data_22[' Pressure (mBar)']=data_22[' Pressure (mBar)'].apply(lambda x: x.replace(',','.')).astype(float)
data_23=pd.read_csv("test_-6_3_tab.txt",delimiter="\t")
data_23.drop(data_23.tail(1).index,inplace=True) 
data_23[' Pressure (mBar)']=data_23[' Pressure (mBar)'].apply(lambda x: x.replace(',','.')).astype(float)
```
